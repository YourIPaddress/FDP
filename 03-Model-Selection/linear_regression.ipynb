# ğŸ¤– The Machine Learning Landscape

From a business perspective, Machine Learning models are generally divided into two strategic categories:

1.  **Supervised Learning** ğŸ¯: (The "Guided" approach) We train the model using historical data that already contains the "correct answers" (labels).
2.  **Unsupervised Learning** ğŸ”: (The "Discovery" approach) We give the model raw data without answers and ask it to find hidden patterns or structures (e.g., customer segmentation).

---

## ğŸ§  Deep Dive: Supervised Learning
In Supervised Learning, we are primarily solving two types of business problems:

### 1. Regression (Forecasting Quantities) ğŸ“ˆ
**"How much?" or "How many?"** ğŸ”¢

Regression models are used when the target variable is **numerical and continuous**. Ideally suited for scenarios involving vast amounts of historical data where the goal is to predict a specific figure.
* **Business Examples:** Predicting next quarter's revenue ğŸ’°, estimating housing prices ğŸ , or forecasting inventory demand ğŸ“¦.

### 2. Classification (Categorizing Outcomes) ğŸ—‚ï¸
**"Which one?" or "Yes/No?"** â“

Classification models are used when the target variable is **categorical**. The model analyzes input data to sort it into distinct, pre-defined buckets or classes.
* **Business Examples:** Predicting if a customer will Churn (Yes/No) ğŸšª, or flagging a transaction as Fraudulent vs. Legitimate ğŸ’³.
* **Key Note ğŸ’¡:** While the final output is a category (e.g., "Churn"), these models actually calculate the **probability** (e.g., "85% risk of churn") behind the scenes before making the final decision.

# 8. ğŸ“ˆğŸ“‰Plots for intuition

### ğŸ“Œ What the Scatter Plot Shows

- Each point represents a real observation (e.g., a sale, customer, or month).

- The x-axis is the input (like advertising spend).

- The y-axis is the output (like monthly sales).

- Weâ€™re trying to find a straight line that best describes how x affects y.

### ğŸ“Œ Why We Draw a Best-Fit Line

- The line represents the general trend.

- It smooths out inconsistencies in the data.

- Helps answer the question:
` â€œIf advertising increases, what happens to sales on average?â€ `

### ğŸ“Œ Interpreting the Line

- Slope: How much y changes when x increases by 1 unit.
Example: If slope = 3, then â‚¹1 extra ad spend â†’ â‚¹3 extra revenue.

- Intercept: Expected value of y when x = 0.

### ğŸ“Œ Error / Residual Visualization

- The vertical distance from each point to the line is the error.

- Linear regression tries to find the line that makes these errors as small as possible.


## 8a. Actual vs Predicted scatter

## 8b. Residuals plot


```python

residuals = y_test - pred
plt.figure(figsize=(6,4))

# Add zorder=1 to draw points first
plt.scatter(pred, residuals, alpha=0.6, zorder=1)

# FIX: Make the horizontal zero line red, thicker, and on top
plt.axhline(0, linestyle="--", color="red", linewidth=1, zorder=5)

plt.xlabel("Predicted")
plt.ylabel("Residual (Actual - Pred)")
plt.title("Residuals vs Predicted")
plt.tight_layout()
plt.show()
```
